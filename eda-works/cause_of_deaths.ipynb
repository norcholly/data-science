```python
columns_to_plot = [col for col in df.columns if col not in ["Year", "Country/Territory", "Code"]]

n_cols = 3
n_rows = math.ceil(len(columns_to_plot) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows), sharex=False)

axes = axes.flatten()

for i, col in enumerate(columns_to_plot):
    colors = ["red" if df[col].iloc[j] <= df[col].iloc[j+1] else "green" for j in range(len(df[col])-1)]

    axes[i].plot(df["Year"], df[col], color='black', alpha=0.3)

    axes[i].scatter(df["Year"][:-1], df[col].iloc[:-1], c=colors, s=50)

    axes[i].set_title(f"{col} over Years")
    axes[i].set_ylabel("Number of Deaths")
    axes[i].grid(True)

for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

for ax in axes:
    ax.set_xlabel("Year")

fig.subplots_adjust(top=0.95, bottom=0.05, hspace=0.4, wspace=0.3)

plt.show()
```

Bu düşüş, özellikle bulaşıcı hastalıkların kontrol altına alınması, doğum ve çocuk sağlığı hizmetlerinin iyileştirilmesi, hijyen ve sanitasyon koşullarının geliştirilmesi ile acil durum yönetimi ve çevresel risklerin azaltılması gibi halk sağlığı önlemlerinin etkili olduğunu göstermektedir.

```python
all_columns = [col for col in df.columns if col not in ["Year", "Country/Territory", "Code"]]

max_values = [df[col].max() for col in all_columns]
min_values = [df[col].min() for col in all_columns]

df_max_min = pd.DataFrame({
    "Disease": all_columns,
    "Max": max_values,
    "Min": min_values
})

df_max_min = df_max_min.sort_values(by="Max", ascending=False)

plt.figure(figsize=(20,8))

plt.bar(df_max_min["Disease"], df_max_min["Min"], color='skyblue', edgecolor='black', label='Min')

plt.bar(df_max_min["Disease"],
        df_max_min["Max"] - df_max_min["Min"],
        bottom=df_max_min["Min"], color='purple', edgecolor='black', label='Max')

plt.xticks(rotation=90)
plt.ylabel("Number of Deaths")
plt.title("Minimum and Maximum Deaths per Year by Disease")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Türkiye’de 1990–2019 dönemi göz önünde bulundurulduğunda, belirli hastalıkların bir yılda kaydedilen maksimum ve minimum ölüm sayıları şu şekilde özetlenebilir:

**Kardiyovasküler hastalıklar bir yılda yaklaşık maksimum 175.000, minimum 110.000 ölümle** en yüksek ölüme yol açan sağlık sorunu olarak öne çıkmaktadır. Bunu yaklaşık **maksimum 100.000, minimum 50.000 ölüm ile Neoplazmlar (anormal hücre artışı / kanserler)** takip etmektedir. **Neonatal (yenidoğan) hastalıklar yaklaşık maksimum 45.000, minimum 10.000 ölüm** ile üçüncü sıradadır. **Kronik solunum hastalıkları ve alt solunum yolu enfeksiyonları her biri yaklaşık 30.000 ölümle** önemli sağlık yükü oluşturmaktadır.

Diğer yandan, Alzheimer ve diğer demans türleri, Diyabet, Kronik böbrek hastalıkları, Çevresel zorlayıcı etkiler (Exposure to Forces of Nature) ve Sindirim sistemi hastalıkları, her biri bir yılda yaklaşık 20.000 ölüm ile orta düzeyde bir risk taşımaktadır. Geri kalan hastalıklar ise yıllık maksimum yaklaşık 10.000 ölüm civarında kalmaktadır.

```python
disease_columns = [col for col in df.columns if col not in ["Year", "Country/Territory", "Code"]]

total_deaths = df[disease_columns].sum()

total_deaths = total_deaths.sort_values(ascending=False)

plt.figure(figsize=(14,8))
plt.bar(total_deaths.index, total_deaths.values, color='purple', edgecolor='blacK')
plt.xticks(rotation=90, ha='right')
plt.ylabel("Number of Deaths")
plt.title("Total Deaths from 1990 to 2019")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

Türkiye’de 1990–2019 dönemi göz önünde bulundurulduğunda, belirli hastalıklar nedeniyle kaydedilen toplam ölüm sayıları şu şekilde özetlenebilir:

**Kardiyovasküler hastalıklar toplam yaklaşık 4.000.000 ölüm** ile en yüksek sağlık yükünü oluşturmuştur. Bunu **yaklaşık 2.000.000 ölüm ile Neoplazmlar (anormal hücre artışı / kanserler)** takip etmektedir. **Kronik solunum hastalıkları ve Neonatal (yenidoğan) hastalıklar toplamda yaklaşık 750.000 ölüm** ile orta düzeyde bir risk teşkil etmektedir.

Diyabet, alt solunum yolu enfeksiyonları, Kronik böbrek hastalıkları, Alzheimer ve diğer demans türleri ile Sindirim sistemi hastalıkları, her biri **toplam yaklaşık 500.000 ölümle** önemli bir sağlık yükü oluşturmaktadır.

Buna ek olarak, Trafik kazaları, Kronik karaciğer hastalıkları gibi sağlık sorunları toplam yaklaşık 250.000 ölüm ile dikkat çekmektedir. Geri kalan hastalıklar ise yaklaşık 100.000 veya daha az toplam ölüm ile sınırlı bir etki göstermektedir. Bazı hastalıkların toplam oldukça düşüktür.
